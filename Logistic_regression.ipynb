## Data Cleaning


```python
bank1=pd.get_dummies(columns=['job','education','poutcome','contact'],data=bank)
```

```python
bank1
```

```python
bank1 = bank1.drop(columns=['day','month','duration','marital'])
```

```python
bank1
```

```python
bank1 = bank1.replace({'y':{'yes':1,'no':0}})       #Replacing categorical data with numerical
```

```python
bank1 = bank1.replace({'default':{'yes':1, 'no':0}})
```

```python
bank1 = bank1.replace({'loan':{'yes':1 , 'no':0}})
```

```python
bank1['housing'].nunique()   #Checking number of attributes in housing column
```

```python
bank1 = bank1.replace({'housing':{'yes':1,'no':0}})
```

```python
bank1.info()     #Checking all data types
```

```python
bank1=bank1.drop(columns=['pdays'])
```

```python
X = bank1.iloc[:,1:]
```

```python
X
```

```python
Y = bank1.iloc[: ,0]
```

```python
Y
```

```python
from sklearn.linear_model import LogisticRegression     #Training model
```

```python
model = LogisticRegression()
```

```python
model.fit(X,Y)
```

```python
y_pred=model.predict(X)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score        #Calculating accuracy score
```

```python
accuracy_score(y_pred,Y)
```

```python
from sklearn.metrics import classification_report       #checking support
```

```python
print(classification_report(y_pred,Y))
```
